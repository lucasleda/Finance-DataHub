Conexao com API e alerta de status da requisição

```python
import requests
from plyer import notification
import pandas as pd

# Função para fazer a requisição e validar a resposta
def fetch_api_data(api_url, api_name):
    response = requests.get(api_url)
    if response.status_code == 200:
        notification.notify(
            title=f'Requisição {api_name} Brasil API',
            message=f"Requisição concluída com sucesso. \nCódigo: {response.status_code}",
            timeout=15
        )
        return response.json(), response.status_code
    else:
        notification.notify(
            title=f'Erro na requisição {api_name} API',
            message=f"Houve um erro ao acessar a API. \nCódigo do erro: {response.status_code}",
            timeout=10
        )
        return None, response.status_code

# URLs das APIs e seus respectivos nomes
api_info = {
    "Banks": "https://brasilapi.com.br/api/banks/v1",
    "Corretoras": "https://brasilapi.com.br/api/cvm/corretoras/v1",
    "PIX": "https://brasilapi.com.br/api/pix/v1/participants"
}

# Extraindo os dados das APIs e convertendo para DataFrames
dataframes = {}
status_codes = {}
for name, url in api_info.items():
    data, status_code = fetch_api_data(url, name)
    if data:
        dataframes[name] = pd.DataFrame(data)
    status_codes[name] = status_code

# Verificando se os DataFrames foram criados corretamente (somente uma vez)
for name, status_code in status_codes.items():
    print(f"Código API {name}: {status_code}")
```
